```python
folder_path = "/content/drive/MyDrive/reviser_patch_predictions"
output_path = "ambiguties_tested.csv"
```

```python
image_files = [f for f in os.listdir(folder_path) if f.endswith(".png")]
data = []
for file in image_files:
    image_path = os.path.join(folder_path, file)
    image = Image.open(image_path).convert("L")
    image_array = np.array(image)
    white_pixels = np.argwhere(image_array >= 150)
    for pixel in white_pixels:
        row_cord = pixel[0]
        col_cord = pixel[1]
        center_pixel_value = image_array[row_cord, col_cord]
        data.append([file, row_cord, col_cord , center_pixel_value])
```

```python
df = pd.DataFrame(data, columns=["File_Name", "Row_Coordinate", "Column_Coordinate", "Center_Pixel_Value"])
```
