```python
import cv2
import numpy as np
import os
import pandas as pd
```

```python
learner1_path = '/content/drive/MyDrive/Learner_1'
od_path = '/content/drive/MyDrive/ODsamples'
reviser_cnn_path = '/content/drive/MyDrive/CORRECTED_images'
```

```python
learner1_image_files = sorted(os.listdir(learner1_path))
od_image_files = sorted(os.listdir(od_path))
reviser_cnn_image_files = sorted(os.listdir(reviser_cnn_path))
```

```python
def accuracy(prediction_path,y_test_path,prediction_image_files,y_test_image_files):
    threshold = 0.1
    accuracies_data = []
    for idx in range(len(prediction_image_files)):
        T = 0
        F = 0
        learner1_image_path = os.path.join(prediction_path, prediction_image_files[idx])
        od_image_image_path = os.path.join(y_test_path, y_test_image_files[idx])
        learner1_image = cv2.imread(learner1_image_path)
        od_image = cv2.imread(od_image_image_path)
        learner_image_gray = cv2.cvtColor(learner1_image, cv2.COLOR_BGR2GRAY)
        od_truth_image_gray = cv2.cvtColor(od_image, cv2.COLOR_BGR2GRAY)
        learner_image_array = np.array(learner_image_gray)
        od_image_array = np.array(od_truth_image_gray)

        for i in range(learner_image_array.shape[0]):
            for j in range(learner_image_array.shape[1]):
                if abs(learner_image_array[i][j] - od_image_array[i][j]) < threshold:
                    T += 1
                else:
                    F += 1
        accuracy = T / (T + F)
        accuracies_data.append((accuracy))
    return accuracies_data
```

```python
learner1_accuracy = accuracy(learner1_path,od_path,learner1_image_files,od_image_files)
reviser_cnn_accuracy = accuracy(reviser_cnn_path,od_path,reviser_cnn_image_files,od_image_files)
```

```python
df = pd.DataFrame({'File_Names':learner1_image_files,'Accuracy in Learner CNN 1':learner1_accuracy,'Accuracy in Reviser CNN':reviser_cnn_accuracy})
```
