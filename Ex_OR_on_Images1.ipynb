```python
folder1_path = 'drive/MyDrive/prediction_1/content/Learner_1'
folder2_path = 'drive/MyDrive/prediction_2/content/Learner_2'
output_folder = 'drive/MyDrive/Learner_ambiguities'
os.makedirs(output_folder, exist_ok=True)
```

```python
folder1_images = sorted(os.listdir(folder1_path))
folder2_images = sorted(os.listdir(folder2_path))
```

```python
folder1_images
```

```python
folder2_images
```

```python
for filename1 in folder1_images:
  for filename2 in folder2_images:
    if '_'.join(filename1.split('_')[:2]) == '_'.join(filename2.split('_')[:2]):
      image1_path = os.path.join(folder1_path, filename1)
      image2_path = os.path.join(folder2_path, filename2)
      with Image.open(image1_path) as image1, Image.open(image2_path) as image2:
            result_image = Image.new('RGB', image1.size)
            pixels_result = result_image.load()
            for x in range(image1.width):
                for y in range(image1.height):
                    pixel1 = image1.getpixel((x, y))
                    pixel2 = image2.getpixel((x, y))
                    xor_r = pixel1 ^ pixel2
                    xor_g = pixel1 ^ pixel2
                    xor_b = pixel1 ^ pixel2
                    pixels_result[x, y] = (xor_r, xor_g, xor_b)
            output_filename = f"amb_{'_'.join(filename1.split('_')[:2])}.png"
            output_path = os.path.join(output_folder, output_filename)
            result_image.save(output_path)
```
