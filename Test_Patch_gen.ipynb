```python
import os
import numpy as np
import pandas as pd
from PIL import Image
```

```python
excel_path = "drive/MyDrive/white_pixel.xlsx"
output_folder = "25_patches"
image_folder = "/content/drive/MyDrive/images"
os.makedirs(output_folder, exist_ok=True)
```

```python
df = pd.read_excel(excel_path)
```

```python
df['File_Name']=df['File_Name'].str.split('_').apply(lambda x:"_".join(x[2:]))
```

```python
for i, row in df.iterrows():
    file_name = row["File_Name"]
    row_coord = row["Row_Coordinate"]
    col_coord = row["Column_Coordinate"]
    image_path = os.path.join(image_folder, file_name)
    image = Image.open(image_path)
    left = col_coord - 12
    upper = row_coord - 12
    right = col_coord + 13
    lower = row_coord + 13

    patch = image.crop((left, upper, right, lower))
    patch_filename = f"image_ambiguity_{file_name.split('_')[1][:3]}_25_{row_coord}_{col_coord}.png"
    patch_path = os.path.join(output_folder, patch_filename)
    patch.save(patch_path)
```
